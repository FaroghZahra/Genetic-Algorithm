# QUEENS PROBLEM

```python
#imports
import numpy as np
import random as rand
import sys
```

```python
#creating population, random filling of board column wise
def fill(sample):
    for i in range(0,8):
        position = rand.randrange(0,8)
        sample[position][i] = 1
```

```python
#evaluate fitness

def fitness(sample):
    badnessQ = np.zeros(8)
    for col in range(0,8):
        for row in range(0,8):
            if (sample[row][col] == 1):
                if(badnessQ[col] == 0):  #no collision found in column cuz filled col wise, check rows
                    for Cols in range(0,8):
                        if(sample[row][Cols] == 1 and Cols != col):
                            badnessQ[col] +=1
                            break
                if(badnessQ[col] == 0): # check lower right diagnol
                    diaLrR = row
                    diaLrC = col
                    while((diaLrC <= 7) and (diaLrR <= 7)):
                        if(sample[diaLrR][diaLrC] == 1 and diaLrC != col and diaLrR != row):
                            badnessQ[col] +=1
                            break
                        diaLrR += 1
                        diaLrC += 1
                if(badnessQ[col] == 0): # check upper right diagnol
                    diaUrR = row
                    diaUrC = col
                    while((diaUrC >= 0) and (diaUrR >= 0)):
                        if(sample[diaUrR][diaUrC] == 1 and diaUrC != col and diaUrR != row):
                            badnessQ[col] +=1
                            break
                        diaUrR -= 1
                        diaUrC -= 1
                if(badnessQ[col] == 0): # check lower left diagnol
                    diaLlR = row
                    diaLlC = col
                    while((diaLlC >= 0) and (diaLlR <= 7)):
                        if(sample[diaLlR][diaLlC] == 1 and diaLlC != col and diaLlR != row):
                            badnessQ[col] +=1
                            break
                        diaLlR += 1
                        diaLlC -= 1
                if(badnessQ[col] == 0): # check Upper left diagnol
                    diaUlR = row
                    diaUlC = col
                    while((diaUlC <= 7) and (diaUlR >= 0)):
                        if(sample[diaUlR][diaUlC] == 1 and diaUlC != col and diaUlR != row):
                            badnessQ[col] +=1
                            break
                        diaUlR -= 1
                        diaUlC += 1              
    badness = 0
    for i in range (0,8):
        badness = badness + badnessQ[i] 
    return int(badness)

```

```python
# check fitness of all samples and store in dict, if badness zero found exit
def foundzero(sample):
    print("Board with badness 0, no Queen Collisions")
    print(sample)
    sys.exit('Desired output found')
    

def checkifzero(sample,fit,i):
    if(fitness(sample) == 0):
        foundzero(sample)
    else:
        fit[i] = (fitness(sample))
    return fit

def fitnessfunc(sample1,sample2,sample3,sample4):
    fit = dict()
    checkifzero(sample1,fit,0)
    checkifzero(sample2,fit,1)
    checkifzero(sample3,fit,2)
    checkifzero(sample4,fit,3)
    print(fit)
    return fit


```

```python
#selecting keys of best two
def selectbest(fit):
    best = min(fit, key = fit.get)
    del fit[best]
    return best

#selecting best two boards
def selectbestboards(fit,sample1,sample2,sample3,sample4):
    best1 = selectbest(fit)
    sampleb1 = np.zeros((8,8))
    if(best1 == 0):
        sampleb1 = sample1
    elif(best1 == 1):
        sampleb1 = sample2
    elif(best1 == 2):
        sampleb1 = sample3
    elif(best1 == 3):
        sampleb1 = sample4
    return sampleb1
```

```python
#getting list
def getlist(sample):
    arrayp = np.zeros(8)
    for col in range(0,8):
        for row in range(0,8):
            if(sample[row][col] == 1):
                arrayp[col] = row
    return arrayp

#crossover
def crossover(list1,list2):
    temp = np.zeros(8)
    for i in range(0,8):
        if(i < 4):
            temp[i] = list1[i]
        else:
            temp[i] = list2[i]
    return temp
```

```python
#mutation
def mutation(list1):
    changepos = rand.randrange(0,8)
    while(True):
        newnum = rand.randrange(0,8)
        if(list1[changepos] == newnum):
            continue
        else:
            list1[changepos] = newnum
            break
    return list1
```

```python
#generating new offsprings
def newoffspring(sample,list1):
    for i in range(0,8):
        pos = int(list1[i])
        sample[pos][i] = 1
    return sample
```

```python
#main

#will print the board if found else will badness of each sample, indexces indicate board number, and  print msg
#wait few seconds for let it process as the loop has to run 2500 times
sample1 = np.zeros((8,8))
sample2 = np.zeros((8,8))
sample3 = np.zeros((8,8))
sample4 = np.zeros((8,8))

fill(sample1)
fill(sample2)
fill(sample3)
fill(sample4)

fit = fitnessfunc(sample1,sample2,sample3,sample4)
sampleb1 = selectbestboards(fit,sample1,sample2,sample3,sample4)
sampleb2 = selectbestboards(fit,sample1,sample2,sample3,sample4)
list1 = np.zeros(8)
list2 = np.zeros(8)
list1 = (getlist(sampleb1))
list2 = (getlist(sampleb2))
listn1 = np.zeros(8)
listn2 = np.zeros(8)
listn1 = crossover(list1,list2)
listn2 = crossover(list2,list1)
listm1 = np.zeros(8)
listm2 = np.zeros(8)
listm1 = mutation(listn1)
listm2 = mutation(listn2)
newsample1 = newoffspring(np.zeros((8,8)),listm1)
newsample2 = newoffspring(np.zeros((8,8)),listm2)


i = 0
while(i < 2500):
    fit = fitnessfunc(sampleb1,sampleb2,newsample1,newsample2)
    sampleb1 = selectbestboards(fit,sampleb1,sampleb2,newsample1,newsample1)
    sampleb2 = selectbestboards(fit,sampleb1,sampleb2,newsample1,newsample1)
    list1 = np.zeros(8)
    list2 = np.zeros(8)
    list1 = (getlist(sampleb1))
    list2 = (getlist(sampleb2))
    listn1 = np.zeros(8)
    listn2 = np.zeros(8)
    listn1 = crossover(list1,list2)
    listn2 = crossover(list2,list1)
    listm1 = np.zeros(8)
    listm2 = np.zeros(8)
    listm1 = mutation(listn1)
    listm2 = mutation(listn2)
    newsample1 = newoffspring(np.zeros((8,8)),listm1)
    newsample2 = newoffspring(np.zeros((8,8)),listm2)
    i+=1

if(i >= 2500):
    print("Time exceeded solution not found")
```
